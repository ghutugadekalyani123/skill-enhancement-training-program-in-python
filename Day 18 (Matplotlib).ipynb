## problem 5
 
 Generate a pie chart for given data

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.2, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

## Problem statement
Using the Matplotlib library, generate a Pie Chart to categorise student marks data in a CSV file into Distinction, First Class, Second Class, Third Class and Failed Categories(based on percentages). Explode the Distinction category.
Generate a CSV containing marks of 2000 students
From the CSV, generate the Pie Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv('datafiles/student.csv')
x=df.values
x=[]
failed=0
d=0
f=0
s=0
t=0
for row in df.values:
    x.append((row[2]))
for i in range(len(x)):    
        if(x[i]>80):
            d=d+1
        elif(x[i]>70) and (x[i]<=80):
            f=f+1
        elif(x[i]>60) and (x[i]<=70):
            s=s+1
        elif(x[i]>50) and (x[i]<=60):
            t=t+1
        else:
            failed=failed+1
labels = 'distinction', 'first', 'second', 'third','fail'
sizes = [d, f, s, t,failed]
explode = (0, 0, 0, 0,0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df= pd.read_csv('datafiles/AppleStore.csv')
df=df[['prime_genre','price','user_rating']]
df=df[df.prime_genre=='Sports']
free=[]
paid=[]
for row in df.values: 
    if(row[1]==0.00):
        free.append(row[2])
    else:
        paid.append(row[2])

x=sum(free)/len(free)
y=sum(paid)/len(paid)

objects = ('PaidAvg', 'FreeAvg')
y_pos = np.arange(len(objects))
performance = [y,x]
#plt.bar(years,income,width,color='Pink')
plt.bar(y_pos, performance,width=0.1, color='pink')
plt.xticks(y_pos, objects)
plt.ylabel('rating')
plt.title('free and paid app rating')
plt.show()

```
